Lets do some complex stuffs using graphical analysis. Lets check among the boarders which sex survived most from drowning and after that we are also going to check the same thing for children (age<15) where 1 = survived and 0= not survived

```python
fig, axs=plt.subplots(1,2)
df[df.Sex=='male'].Survived.value_counts().plot(kind='barh',ax=axs[0],title='Male')
df[df.Sex=='female'].Survived.value_counts().plot(kind='barh',ax=axs[1],title='Female')
```

```python
fig, axs=plt.subplots(1,2)
df[(df.Age<15)&(df.Sex=='male')].Survived.value_counts().plot(kind='barh',ax=axs[0],title='Male')
df[(df.Age<15)&(df.Sex=='female')].Survived.value_counts().plot(kind='barh',ax=axs[1],title='Female')
```

Lets now do some data analysis taking into account Passenger Class dataset. We are going to see a new function GROUPBY in this section. 

Intro to group by: http://pandas-docs.github.io/pandas-docs-travis/groupby.html

So we will perform :

<li>1.We are going to see total number of people in each class numerically and graphically
<li>2.Check wheather there is any null vaue in the Pclass column/dataset
<li>3.Check which class has more female/male
<li>4.Check the average number of people survived from each class

```python
df.Pclass.value_counts()
```

```python
df.Pclass.hist(bins=15) 
```

```python
df[df.Pclass.isnull()]
```

```python
#Checking distribution of male and female passengers in each Passenger CLass
fig, axs=plt.subplots(1,2)
df[df.Sex=='male'].Pclass.value_counts().plot(kind='barh',ax=axs[0],title='Male Passengers')
df[df.Sex=='female'].Pclass.value_counts().plot(kind='barh',ax=axs[1],title='Female Passengers')
```

```python
#Same as above but now using groupby function
df.groupby(['Sex','Pclass']).PassengerId.count().plot(kind='barh')
```

```python
# Calculating average number of people survived from each class
# Instead of this complicated statement below we can also use : df.groupby('Pclass').Survived.mean()
# df.groupby('Pclass').Survived.mean() returns same result as below statement

((df.groupby('Pclass').Survived.sum()/df.Pclass.value_counts())*100).plot(kind='bar')
```
