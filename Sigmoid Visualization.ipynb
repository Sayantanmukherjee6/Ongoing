```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
import math

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
```

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```

### Without Bias term

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(np.dot(x,0.55))
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(np.dot(x,1))
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(np.dot(x,2))
plt.plot(x,sig)
plt.show()
```

### With Bias term

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(np.dot(x,1)-2)
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)
sig = sigmoid(np.dot(x,1)-5)
plt.plot(x,sig)
plt.show()
```

### Increasing Feature complexity

```python
# Weights need to change
x = np.arange(-10., 10., 0.2)**2
sig = sigmoid(np.dot(x,0.1)-5)
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)**10
sig = sigmoid(np.dot(x,0.000000003)-10)
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)*2
sig = sigmoid(np.dot(x,1)-5)
plt.plot(x,sig)
plt.show()
```

```python
x = np.arange(-10., 10., 0.2)*10
sig = sigmoid(np.dot(x,0.2)-5)
plt.plot(x,sig)
plt.show()
```
